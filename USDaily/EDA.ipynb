```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
```

```python
def read_dir(directory):
    dfs = []
    for file in tqdm(os.listdir(directory)):
            try:
                df = pd.read_csv(os.path.join(directory, file))
            except:
                continue
            ticker = file.split('.')[0].upper()
            df['Ticker'] = ticker
            df.set_index('Ticker', inplace=True)
            dfs.append(df)
        df = pd.concat(dfs)
    return df.sort_values('Date')

def read_market_data(stocks=True, etfs=False, stocks_dir='Data/Stocks', etfs_dir='Data/ETFs'):
    dfs = []
    if stocks:
        print('...Reading in Stocks Data...')
        stocks = read_dir(stocks_dir)
    if etfs:
        print('...Reading in ETFs Data...')
        etfs = read_dir(etfs_dir)
    return stocks, etfs

df = read_market_data()
```
